```python
import pandas as pd
# read the data
url="https://raw.githubusercontent.com/GabrieleKasputyte/Duomenu-gavyba/main/CSV/mortality1.csv"

data = pd.read_csv(url)
data.head()
```

```python
data['Year'] = data['Year'].astype(object)
data.describe()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
# Plot sepal width as a function of sepal_length across days
g = sns.lmplot(
    data=data,
    x="Population", y="RSuicBoth", hue="Sex",
    height=5
)

# Use more informative axis labels than are provided by default
g.set_axis_labels("Population", "Suicides no.")
```

```python
sns.relplot(
    data=data, kind="line",
    x="Year", y="Rsuic", hue = "Sex"
)
```

```python
salis = data[(data['Country']=="Lithuania")]
salis
```

```python
sns.set_theme(style="dark")
flights = sns.load_dataset("flights")


# Plot each year's time series in its own facet
g = sns.relplot(
    data=salis,
    x="Population", y="RSuicBoth", col="Year", hue="Year",
    kind="line", palette="crest", linewidth=4, zorder=5,
    col_wrap=3, height=2, aspect=1.5, legend=False,
)

# Iterate over each subplot to customize further
for year, ax in g.axes_dict.items():

    # Add the title as an annotation within the plot
    ax.text(.8, .85, year, transform=ax.transAxes, fontweight="bold")

    # Plot every year's time series in the background
    sns.lineplot(
        data=salis, x="Population", y="RSuicBoth", units="Year",
        estimator=None, color=".7", linewidth=1, ax=ax,
    )

# Reduce the frequency of the x axis ticks
ax.set_xticks(ax.get_xticks()[::2])

# Tweak the supporting aspects of the plot
g.set_titles("")
g.set_axis_labels("", "Suicides")
g.tight_layout()
```
